```python
import nltk, re, string, collections
nltk.download('punkt')
import numpy as np
import random
import string
import time
from pathlib import Path
from collections import OrderedDict
from itertools import chain 


novelFlag = False

while(novelFlag == False):
    print("Choose a Novel Number accordingly:\n")
    print("1-BİLİM İŞ BAŞINDA\n")
    print("2-BOZKIRDA\n")
    print("3-DEĞİŞİM\n")
    print("4-DENEMELER\n")
    print("5-UNUTULMUŞ DİYARLAR\n")
    novelNum = input("Please enter a number correctly:\n")
    novelNum = int(novelNum)
    novelFlag = False
    if(novelNum == 1):
        novelFlag = True
        novelName = 'BİLİM İŞ BAŞINDA'
        print('Selected novel is: {0}'.format(novelName))
        
    elif(novelNum == 2):
        novelFlag = True
        novelName = 'BOZKIRDA'
        print('Selected novel is: {0}'.format(novelName))
        
    elif(novelNum == 3):
        novelFlag = True
        novelName = 'DEĞİŞİM'
        print('Selected novel is: {0}'.format(novelName))
        
    elif(novelNum == 4):
        novelFlag = True
        novelName = 'DENEMELER'
        print('Selected novel is: {0}'.format(novelName))
        
    elif(novelNum == 5):
        novelFlag = True
        novelName = 'UNUTULMUŞ DİYARLAR'
        print('Selected novel is: {0}'.format(novelName))
        
    else:
        novelFlag = False

def getNGrams(wordlist, n):#this function helps me to create n-grams.
    ngrams = []
    for i in range(len(wordlist)-(n-1)):
        ngrams.append(wordlist[i:i+n])
    return ngrams


if(novelFlag == True):#if we read a novel, then we can read it.
    n = input("Please enter a number(n) to apply n-gram algorithm:\n")
    n = int(n)
    start_time = time.time()#start the time.
    text = Path(novelName + '.txt').read_text()#read the selected novel to a string.
    punctuationNoPeriod = "[" + re.sub("\.","",string.punctuation) + "]"#remove punctuations except dot.
    text = re.sub(punctuationNoPeriod, "", text)
    
    tokenized = text.split()#split the text.
    n_grams = getNGrams(tokenized, n)#get ngrams with the help of the getNGrams function.
    ngram_count = collections.Counter(map(tuple, n_grams))#convert this into a tuple, then create a list with appropriate counts.
    sorted_ngram = sorted(ngram_count.items(), key=lambda x: x[1], reverse=True)#sort this by counts. it's from reverse because list is ascending.
    od = OrderedDict(sorted_ngram)#with this sorted, create an ordered dictionary.
    if(n == 1):#get ngram methode that we will use.
        nGramMethode = 'Unigram'
    elif(n == 2):
        nGramMethode = 'Bigram'
    elif(n == 3):
        nGramMethode = 'Trigram'
    else:
        nGramMethode = '{0}-gram'.format(n)
    i = 1#print this ordered dictionary that we just created.
    print('Top 100 words with counters for novel named: {0} with {1} algorithm:'.format(novelName,nGramMethode))
    for key, value in od.items():
        stringKey = str(key)
        stringKey = stringKey.replace(',' , "").replace('(', "").replace(')', "").replace("' '", " ")
        print('{0} - {1}--> {2}'.format(i,stringKey,value))
        i = i + 1
        if i == 101:
            break
            
elapsed_time = time.time() - start_time#get the total elapsed time.
print('Total Elapsed Time:{0}'.format(elapsed_time))#print that too.
```
